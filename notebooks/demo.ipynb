# Sovereign — environment walk-through

A short notebook that builds the env, plays a random episode, prints the ASCII
render, and plots θ and L over time. Useful as a sanity check after install.

```python
import numpy as np
import matplotlib.pyplot as plt
from sovereign import SovereignEnv

env = SovereignEnv()
obs, info = env.reset(seed=0)
print('observation dim:', obs.shape)
print('action n:', env.action_space.n)
```

```python
thetas, legitimacies, rewards = [], [], []
for t in range(env.cfg.limits.t_max):
    a = int(env.action_space.sample())
    obs, r, term, trunc, info = env.step(a)
    thetas.append(info['theta'])
    legitimacies.append(info['legitimacy'])
    rewards.append(r)
    if term or trunc:
        print('terminal_reason:', info['terminal_reason'])
        break

print('total return:', sum(rewards))
```

```python
fig, ax = plt.subplots(2, 1, figsize=(8, 4), sharex=True)
ax[0].plot(thetas, label=r'$\theta$')
ax[0].axhline(0.6, color='orange', linestyle='--', linewidth=0.8)
ax[0].axhline(-0.6, color='green', linestyle='--', linewidth=0.8)
ax[0].set_ylim(-1, 1)
ax[0].legend()
ax[1].plot(legitimacies, color='C1', label='L')
ax[1].set_ylim(0, 1)
ax[1].legend()
ax[1].set_xlabel('turn')
plt.tight_layout()
plt.show()
```
